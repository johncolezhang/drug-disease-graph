```python
import pandas as pd
```

```python
df_all_drug = pd.read_csv("d:/pgkb_graph/processed/all_drug.csv", dtype=str).fillna("")
df_pgx_drug = pd.read_csv("d:/pgkb_graph/processed/pgx_drug_stats.csv", dtype=str).fillna("")
df_pgx_drug_clinical = pd.read_csv("d:/pgkb_graph/processed/pgx_drug_clinical_stats.csv", dtype=str).fillna("")
df_pgx_drug_clinical_1A_1B = pd.read_csv("d:/pgkb_graph/processed/pgx_drug_clinical_stats_1A_1B.csv", dtype=str).fillna("")
df_pgx_drug_research = pd.read_csv("d:/pgkb_graph/processed/pgx_drug_research_stats.csv", dtype=str).fillna("")
df_pgx_drug_organization_label = pd.read_csv("d:/pgkb_graph/processed/pgx_drug_label_stats.csv", dtype=str).fillna("")
df_pgx_drug_fda_label = pd.read_csv("d:/pgkb_graph/processed/pgx_fda_drug_label_stats.csv", dtype=str).fillna("")
```

```python
df_translation = pd.read_csv("processed/chemical_translation.csv", dtype=str).fillna("")
translation_dict = dict(zip(list(df_translation["chemical_name"].str.strip().str.lower().values), 
                            list(df_translation["cn_chemical_name"].str.strip().str.lower().values)))
```

```python
def en_to_cn(en_str):
    return "; ".join([translation_dict[x.strip()] for x in en_str.split(";")])

df_pgx_drug = df_pgx_drug.assign(
    cn_chemical_name=df_pgx_drug.apply(
        lambda row: en_to_cn(row["chemical_name"].strip().lower()), axis=1))
df_pgx_drug = df_pgx_drug[["drug_name", "chemical_name", "cn_chemical_name"]]
```

```python
# add is_pgx_drug column
df_merge = pd.merge(df_all_drug, df_pgx_drug, how="left", on=["drug_name"]).fillna("")
df_merge = df_merge.assign(is_pgx_drug=df_merge.apply(lambda row: "Yes" if row["chemical_name"] != "" else "No", axis=1))
```

```python
# add is_pgx_clinical_drug column
df_merge = pd.merge(df_merge, df_pgx_drug_clinical, how="left", on=['drug_name']).fillna("")
df_merge = df_merge.assign(is_pgx_clinical_drug=df_merge.apply(lambda row: "Yes" if row["clinical_relation_count"] != "" else "No", axis=1))
df_merge = df_merge[["drug_name", "chemical_name", "cn_chemical_name", "is_pgx_drug", 
                     "is_pgx_clinical_drug"]]
```

```python
# add is_pgx_clinical_drug_1A_1B column
df_merge = pd.merge(df_merge, df_pgx_drug_clinical_1A_1B, how="left", on=['drug_name']).fillna("")
df_merge = df_merge.assign(is_pgx_clinical_drug_1A_1B=df_merge.apply(lambda row: "Yes" if row["clinical_relation_count"] != "" else "No", axis=1))
df_merge = df_merge[["drug_name", "chemical_name", "cn_chemical_name", "is_pgx_drug", 
                     "is_pgx_clinical_drug", "is_pgx_clinical_drug_1A_1B"]]
```

```python
# add is_pgx_research_drug column
df_merge = pd.merge(df_merge, df_pgx_drug_research, how="left", on=['drug_name']).fillna("")
df_merge = df_merge.assign(is_pgx_research_drug=df_merge.apply(lambda row: "Yes" if row["research_relation_count"] != "" else "No", axis=1))
df_merge = df_merge[["drug_name", "chemical_name", "cn_chemical_name", "is_pgx_drug",
                     "is_pgx_clinical_drug", "is_pgx_clinical_drug_1A_1B", "is_pgx_research_drug"]]
```

```python
# add is_organization_label column
df_merge = pd.merge(df_merge, df_pgx_drug_organization_label, how="left", on=['drug_name']).fillna("")
df_merge = df_merge.assign(is_organization_label=df_merge.apply(lambda row: "Yes" if row["label_relation_count"] != "" else "No", axis=1))
df_merge = df_merge[["drug_name", "chemical_name", "cn_chemical_name", "is_pgx_drug",
                     "is_pgx_clinical_drug", "is_pgx_clinical_drug_1A_1B", "is_pgx_research_drug",
                     "is_organization_label"]]
```

```python
# add is_fda_label column
df_merge = pd.merge(df_merge, df_pgx_drug_fda_label, how="left", on=['drug_name']).fillna("")
df_merge = df_merge.assign(is_fda_label=df_merge.apply(lambda row: "Yes" if row["label_relation_count"] != "" else "No", axis=1))
df_merge = df_merge[["drug_name", "chemical_name", "cn_chemical_name", "is_pgx_drug",
                     "is_pgx_clinical_drug", "is_pgx_clinical_drug_1A_1B", "is_pgx_research_drug",
                     "is_organization_label", "is_fda_label"]]
```

```python
df_merge.to_csv("processed/all_drug_stats.csv", index=False)
```
