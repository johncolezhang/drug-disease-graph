```python
folder = "C:/Users/zhangke1/Desktop/google-scholar-crawler/output"
fn_list = [os.path.join(folder, x) for x in os.listdir(folder)]

columns = ["title", "content", "url", "sub_title", "year"]
df_all_paper = pd.DataFrame(columns=columns)

for fn in fn_list:
    df_all_paper = pd.concat([df_all_paper, pd.read_csv(fn, dtype=str).fillna("")], axis=0, ignore_index=True)
```

```python
df_all_paper = df_all_paper[(df_all_paper["title"] != "") & (df_all_paper["content"] != "")]
df_all_paper = df_all_paper.drop_duplicates("title")
```

```python
def remove_punctuation(s_in, s_to=u' '):
    if isinstance(s_in, str):
        not_letters = u'!"#%\'()*+,-./:;<=>?@[\]^_`{|}~\n'
        trans_table = dict((ord(char), s_to) for char in not_letters)
        return s_in.translate(trans_table)
    else:
         return s_in.translate(string.maketrans("", ""), string.punctuation)

def parse_year(text):
    pattern = re.compile(r'\s[1-2][0|8|9][0-9]{2}\s')
    year = pattern.findall(text)

    if year:
        year = remove_punctuation(year[0])
        year = year.strip()
        try:
            year = int(year)
            return year
        except:
            year = 0
            return year
    else:
        year = 0
        return year
```

```python
df_all_paper["year"] = [parse_year(x) for x in df_all_paper["sub_title"].values]
```

```python
def check_bgi(row):
    bgi_list = ["bgi", ""]
    if any([True if bgi in row["title"].lower() else False for bgi in bgi_list]) \
       or any([True if bgi in row["content"].lower() else False for bgi in bgi_list]) \
       or any([True if bgi in row["sub_title"].lower() else False for bgi in bgi_list]):
        return True
    
    try:
        res = requests.get(
            row["url"], 
            headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:92.0) Gecko/20100101 Firefox/92.0'},
            timeout=2,
            verify=False
        )
    except:
        return False
    
    if any([True if bgi in res.text.lower() else False for bgi in bgi_list]):
        return True
    return False

bgi_involved_list = []

for index, row in df_all_paper.iterrows():
        bgi_involved_list.append(check_bgi(row))
```
